# Considering tendency budgets in regridded vertical coordinates

_11/10/19_ A few summarizing thoughts. The online rebinning conserves the depth-integrated budget to machine precision. This is to be expected given the way that the remapping is performed online. However, the budget is not closed (errors are substantial) for each layer in the remapped grid (Section 1).  

A comparison of tendencies on the native, remapped, and offline regridded grids (Sections 2 and 3) shows that remapping introduces considerable errors, while offline regridding is generally closer but has missing values (where the binning procedure has not found values). It is not clear to me what is the best way to do a "direct" comparison, i.e. what to use as the vertical coordinate. At present I have used the mean rho2 within each vertical layer, irrespective of the grid. I still need to consider whether this is the most approriate way to compare the grids, for example it could be "double-counting" the impact of remapping (on both the vertical coordinate and the tendency).  

Offline binning conserves the tendency budget exactly (Section 4), both in the vertical sum and in each layer ("bin"). I wonder if this is always the case? For example, if the spacing of the bins excludes some layers present in the native grid. This should be avoidable as long as the outer bounds of the bins are outside the min and max of the variable in the native grid.

### 3. Compare native, online remapped, and offline binned (xhistogram) terms  
Consider how individuals terms are impacted by "binning" data offline using histogram.  
Does the binning approach reproduce something comparable to the online remapping.

```python
# xhistogram binning of model data, nabbed from jbusecke

def vertical_rebin(data, bin_data, bins, dz, vert_dim="st_ocean"):
    nanmask = np.isnan(data)
    # Should we also check the bin data for nans?
    full_sum = histogram(
        bin_data.where(~nanmask),
        bins=[bins],
        weights=(data * dz).where(~nanmask),
        dim=[vert_dim],
    )
    return full_sum

def vertical_rebin_wrapper(
    ds,
    bin_data_name,
    bins,
    dz_name="dz",
    vert_dim="st_ocean",
    return_average=True,
    debug=False,
):
    """A wrapper for the core functionality in `vertical_rebin`.
    Accepts datasets and calculates the average over the new depth coordinates.
    """
    ds = ds.copy()
    ds_rebinned = xr.Dataset()
    
    ones = xr.ones_like(ds[dz_name])
    
    dz_rebinned = vertical_rebin(
        ones,
        ds[bin_data_name],
        bins,
        ds[dz_name],
        vert_dim=vert_dim,
    )
    for var in ds.data_vars:
        if ds[var].dtype == 'float':
            ds_rebinned[var] = vertical_rebin(
                ds[var], ds[bin_data_name], bins, ds[dz_name], vert_dim=vert_dim
            )
    if return_average:
        ds_rebinned = (
            ds_rebinned / dz_rebinned
        )  # this might cause a lot of overhead...i can try to deactivate if the save fails.

    ds_rebinned[dz_name] = dz_rebinned

    return ds_rebinned
```

```python
prof_native_rebinned = vertical_rebin_wrapper(prof_native,"rhopot2",prof_rho2.rho2_i.values,dz_name="thkcello",vert_dim="zl")
```
